```python
import os
from ultralytics import YOLO
```

```python
model=YOLO('yolov8n.pt')
```

```python
res=model.train(data=os.path.join(ROOT_DIR,'data.yaml'),epochs=100)
```

```python
model.save('cnic_text_detection')
```

```python
model.val()
```

```python
import matplotlib.pyplot as plt
```
